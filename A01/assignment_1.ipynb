```python
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# HF model import
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained(
    "Lykon/dreamshaper-8",
    dtype=torch.float16,
    ).to(device)
```

```python
# Absurd prompts
prompts = [
    "A surreal collage-style scene inspired by Monty Python humor:\
    a medieval knight in full armor sits at a tiny office desk in an endless grassy field,\
    seriously arguing with a giant chicken wearing a hat.\
    Absurd, witty, deadpan comedy, dreamlike logic, symbolic stire,\
    vintage collage aesthetic, cut-paper texture, muted colors.",

    "A surreal collage-style scene inspired by Monty Python humor:\
    a medieval knight in full armor sits at a tiny office desk in an desert's oasis,\
    stamping paperwork while crocodyles are dancing around his desk.\
    Absurd, witty, deadpan comedy, dreamlike logic, symbolic stire,\
    vintage collage aesthetic, cut-paper texture, muted colors.",

    "A surreal collage-style scene inspired by Monty Python humor:\
    a medieval knight in full armor sits at a tiny office desk in a stary night,\
    watching elephants with tiny hat walking by him.\
    Absurd, witty, deadpan comedy, dreamlike logic, symbolic stire,\
    vintage collage aesthetic, cut-paper texture, muted colors."
]
```

```python
for i, prompt in enumerate(prompts, start=1):

    image = pipe(prompt).images[0]

    image.save(f"/content/drive/MyDrive/surreal_collage_{i}.png")

    with open(f"/content/drive/MyDrive/surreal_collage_{i}.txt", "w") as f:
        f.write(prompt)

    print(f"Saved surreal_collage_{i}.png and .txt")
```
