```python
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
```

```python
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained(
    "dreamlike-art/dreamlike-photoreal-2.0",
    torch_dtype=torch.float16
    ).to(device)
```

```python
prompt = "Surrealist collage in Monty Python-style, featuring a head rising from the ground, a dog with a hat, and confetti."

image = pipe(prompt).images[0]
image.save("surreal_collage_1.jpg")

prompt = "Surrealist collage in Monty Python-style, featuring a head in the sky, a cat playing checkers, and flying clocks."

image = pipe(prompt).images[0]
image.save("surreal_collage_2.jpg")

prompt = "Surrealist collage in Monty Python-style, featuring hands with eyes, a crocodile with a suit, and flowers with mouths."

image = pipe(prompt).images[0]
image.save("surreal_collage_3.jpg")
```
