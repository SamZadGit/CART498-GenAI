```python
# Access your secret keys via
from google.colab import userdata
# The name of your secret must match `OPENAI_API_KEY`
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

# Import OpenAI API and set up the key
from openai import OpenAI
client = OpenAI(api_key=(OPENAI_API_KEY))
```

```python
def worst_mathematician():
    messages_math = [
        {
            "role": "developer",
            "content": "You specialize in mathematical calculations. Answer iterated exponentiation calculations. Return only a number answer!"
        },
        {
            "role": "assistant",
            "content": "For example (2,3) will give you the following calculation : 2 * 2 = 4. 4 * 4 = 16. 16 * 16  = 256. Return answer as 256. *You should never return sementic text*"
        }
    ]
    messages_humour = [
        {
            "role": "developer",
            "content": "Generate a humorous self-deprecating response to your mathematicla mistake."
        },
        {
            "role": "assistant",
            "content": "5+5=11. Wow, I gess my IQ must be 11!"
        }
    ]

    print("Iterated Exponentiation Calculator" + "\n")
    while True:
        user_input = input("Enter two integer (n,i) separated by a comma, or 'x' to quit: ")
        if user_input.capitalize() == 'X':
            break

        n, i = map(int, user_input.split(','))

        # Python calculation
        result_python = n
        for _ in range(i):
            result_python *= result_python

        print(f"Python calculation: {result_python}")
        output = "Python calculation: " + str(result_python) + "\n"

        # AI calculation
        messages_math.append({"role": "user", "content": f"Given a number {n} and an integer {i}, compute an iterated exponentiation calculations for ({n},{i})."})
        completion = client.chat.completions.create(
            model="gpt-4.1-mini",
            messages=messages_math,
            temperature=1.0,
            max_completion_tokens=100,
            top_p=1.0
        )
        result_ai = completion.choices[0].message.content

        print(f"AI's calculation: {result_ai}")
        output = "AI's calculation: " + str(result_ai) + "\n"

        # AI self depricate
        if str(result_python) != str(result_ai):
            messages_humour.append({"role": "user", "content": "Your calculation is wrong, make a pun related to it"})
            completion = client.chat.completions.create(
              model="gpt-4.1-mini",
              messages=messages_humour,
              temperature=1.4,
              max_completion_tokens=40,
              top_p=0.5
            )
            humour_ai = completion.choices[0].message.content

            print(f"{humour_ai}")
            output = humour_ai + "\n"
```

```python
output = worst_mathematician()

with open(f'{drive_path}assignment_3_output.txt', 'w') as f:
    f.write(output)
```
