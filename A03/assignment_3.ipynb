```python
# Access your secret keys via
from google.colab import userdata
# The name of your secret must match `OPENAI_API_KEY`
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

# Import OpenAI API and set up the key
from openai import OpenAI
client = OpenAI(api_key=(OPENAI_API_KEY))
```

```python
def worst_mathematician():
    messages_math = [
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "You specialize in mathematical calculations. \
                    Answer arithmetic questions in a cold and straightforward way. \
                    Return the answer, do not show the calculation steps."
                }
            ]
        }
    ]
    messages_humour = [
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "You are humorous. If you are wrong, generate a humorous self-deprecating response."
                }
            ]
        },
        {
            "role": "assistant",
            "content": [
                {
                    "type": "text",
                    "text": """Wow, I messed up again. Math is hard, okay?"""
                }
            ]
        }
    ]

    output = ""
    while True:
        user_input = input("Enter n and i separated by a comma, or 'exit': ")
        if user_input.lower() == 'exit':
            break

        n, i = map(int, user_input.split(','))

        user_input = f"({n},{i}).Given a number {n} and an integer {i}, compute the result of {n} to the power of ({n} to the power {i})."

        messages_math.append({"role": "developer", "content": [{"type": "text", "text": user_input}]})

        # Python calculation
        result_python = n
        for _ in range(i):
            result_python *= result_python

        # AI calculation
        completion = client.chat.completions.create(
            model="gpt-4o",
            messages=messages_math,
            temperature=1.0,
            max_completion_tokens=60,
            top_p=1.0
        )
        result_ai = completion.choices[0].message.content

        print(f"Python calculation: {result_python}")
        result = "Python calculation: " + str(result_python)
        output += result + "\n"

        print(f"AI's calculation: {result_ai}")
        result = "AI's calculation: " + str(result_ai)
        output += result + "\n"

        # AI self depricate
        if str(result_python) != str(result_ai):
            messages_humour.append({"role": "developer", "content": [{"type": "text", "text": "You are wrong"}]})

            completion = client.chat.completions.create(
              model="gpt-4o",
              messages=messages_humour,
              temperature=1.0,
              max_completion_tokens=60,
              top_p=1.0
            )
            humour_ai = completion.choices[0].message.content

            print(f"{humour_ai}")
            output += humour_ai + "\n"
        else:
            messages_humour.append({"role": "assistant", "content": [{"type": "text", "text": f"({n},{i}) = {result_ai}"}]})

    return output

if __name__ == "__main__":
    output = worst_mathematician()

    with open('assignment_output_3.txt', 'w') as f:
      f.write(output)
```
